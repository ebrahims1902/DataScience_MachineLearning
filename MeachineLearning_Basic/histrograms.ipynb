# Histrograms

Working with histrogram we want only single dimsation Arry 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_suger_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_suger_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel('Suger range')
plt.ylabel('Total no of patients')
plt.title('Blood Suger Analysis')

plt.hist([blood_suger_men,blood_suger_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```
