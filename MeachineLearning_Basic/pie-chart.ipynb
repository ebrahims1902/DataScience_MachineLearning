```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","car","Other Utilities"] 
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',explode=[0,0.1,0.1,0,0],startangle=180)
plt.show()
```
